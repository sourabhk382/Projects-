```python
q5=df['Q5'][1:]
role_software=pd.concat([q5,softwares],axis=1)
```

```python
role_software_binary=role_software[1:]
```

```python
role_software_sum=role_software_binary.groupby(['Q5'])[['Q7_Part_1','Q7_Part_2','Q7_Part_3','Q7_Part_4','Q7_Part_5','Q7_Part_6','Q7_Part_7','Q7_Part_8','Q7_Part_9','Q7_Part_10','Q7_Part_11',
      'Q7_Part_12','Q7_OTHER']].apply(lambda x: x.astype(int).sum())
```

```python
role_software_sum.rename(columns={'Q7_Part_1':'Python'}, inplace=True)
role_software_sum.rename(columns={'Q7_Part_2':'R'}, inplace=True)
role_software_sum.rename(columns={'Q7_Part_3':'SQL'}, inplace=True)
role_software_sum.rename(columns={'Q7_Part_4':'C'}, inplace=True)
role_software_sum.rename(columns={'Q7_Part_5':'C++'}, inplace=True)
role_software_sum.rename(columns={'Q7_Part_6':'Java'}, inplace=True)
role_software_sum.rename(columns={'Q7_Part_7':'Javascript'}, inplace=True)
role_software_sum.rename(columns={'Q7_Part_8':'Julia'}, inplace=True)
role_software_sum.rename(columns={'Q7_Part_9':'Swift'}, inplace=True)
role_software_sum.rename(columns={'Q7_Part_10':'Bash'}, inplace=True)
role_software_sum.rename(columns={'Q7_Part_11':'MATLAB'}, inplace=True)
role_software_sum.rename(columns={'Q7_Part_12':'None'}, inplace=True)
role_software_sum.rename(columns={'Q7_OTHER':'Other'}, inplace=True)
```

```python
role_software_sum
```

```python
role_software_percent=role_software_sum.join(rolecount)
#role_software_sum['total']=role_software_sum.sum(axis=1)                   #sum of row
role_software_percent=role_software_percent.rename(columns={'Q5':'Total'})
```

```python
role_software_percentage = role_software_percent.div(role_software_percent['Total'], axis=0)*100
role_software_percentage=role_software_percentage.drop(['Total'],axis=1)
```

```python
plt.figure(figsize=(30,12))                                                                     #heatmap
plt.rcParams.update({'font.size':25})
sns.heatmap(data=role_software_percentage, annot=True,cmap='Blues') 
plt.xlabel('Programming languages',fontsize=30,fontweight='bold',labelpad=25)
plt.ylabel("Role",fontsize=30,fontweight='bold',labelpad=25)
plt.title("Kaggle survey Rolewise Programming Languages",fontsize=35,fontweight='bold',y=1.02)
```

```python
role_software_python=role_software_percentage.sort_values(by=['Python'],ascending=False)
```

```python
plt.figure(figsize=(25,10))
plt.rcParams.update({'font.size':20})
sns.barplot(y=role_software_python.index,x=role_software_python['Python'], palette=("Blues_r")) 
plt.xlabel("% users") 
plt.ylabel("Role")
plt.title("Kaggle survey Role wise Python users (percent)")
```
