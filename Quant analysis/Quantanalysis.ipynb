```python
from nsepy import get_history
import pandas as pd   
import numpy as np           
from nsepy import get_index_pe_history
pd.plotting.register_matplotlib_converters()           
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns

from datetime import date
import yfinance as yf
```

```python
nifty_pe = get_index_pe_history(symbol="NIFTY",
                                start=date(2011,1,1),
                                end=date(2021,11,30)) 
nifty_pe.tail()
```

```python
plt.figure(figsize=(25,10))
sns.set_style('whitegrid')
plt.rcParams.update({'font.size': 30})
sns.lineplot(data=nifty_pe['P/E'])
```

```python
nifty_pe.describe()   
```

```python
plt.figure(figsize=(10,5))
sns.set_style('whitegrid') 
plt.rcParams.update({'font.size': 15}) 
plt.hist(nifty_pe['P/E'], bins=20)
plt.title('Nifty P/E distribution')
```

```python
sbi = get_history(symbol="SBIN", start=date(2016,1,1), end=date(2021,11,30)) 
plt.rcParams.update({'font.size': 15})
sbi[['Close']].plot()
```

```python
sbi.tail() 
```

```python
sbi_return = round(np.log(sbi.Close).diff() * 100, 2)
sbi_return.dropna(inplace = True)
sbi_return.tail()
```

```python
sbi_return.describe()
```

```python
plt.figure(figsize=(12,7))
sns.set_style('whitegrid') 
plt.rcParams.update({'font.size': 15}) 
plt.hist(sbi_return, bins=50)
plt.title('sbi return distribution')
```
